```python
import torch 
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("vinai/phobert-base")
```

```python
lst_all = []
```

```python
with open("/content/corpus.txt", "r") as f:
  f_contents = f.readlines()
  lst_all.append(f_contents)
```

```python
ques = []
```

```python
ans = []
```

```python
len(lst_all[0])
```

```python
for i in range(0, len(lst_all[0])-1):
    a = lst_all[0][i].split("<0000>")
    ques.append(a[0])
    ques[i] = ques[i].strip()
    ans.append(a[1])
```

```python
count = 0
top_k = []

for i in range(len(cosine[0])):
  if cosine[0][i] > 0.8:
    count += 1
    top_k.append(cosine[0][i])  
    print("Question: ", ques[i])
    print("Answer: ", ans[i])
```
